```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##Criar figura e eixos ###
fig, ax = plt.subplots()
nomefig = 'teste1.png'
#############################
ax.plot([2,3,4])

## Salvar a figura
fig.savefig(nomefig)
plt.show
```

```python
### estrutura de dados como array
fig, ax = plt.subplots()
nomefig = 'teste2.png'

x = np.array([2,3,4,7,10,11])
ax.plot(x, x**2) ##aparece uma parÃ¡bola
fig.savefig(nomefig)
plt.show()

```

```python
### estrutura de dados como array
fig, ax = plt.subplots()
nomefig = 'teste3.png'

x = np.array([2,2.5, 3, 3.5,4,7,10,11])
x = np.linspace(0, 15, 100)
ax.plot(x, np.sin(x)) ##aparece uma senoide
        
fig.savefig(nomefig)
plt.show()
```

```python
### estrutura de dados como array
fig, ax = plt.subplots(figsize = (9, 6))
nomefig = 'teste4.png'

##definindo variaveis tamanho da letra
xlab = 'Tempo em relaÃ§Ã£o a quinta-feira 11/08/2021' #nome da label
fsize = 8 #tamanho da letra
##($\eta $C) aparece um eta, ($\int $C) aparece uma integral,  ($^o$C )/($\mathrm{^o}$C) graus CÃ©lsios
ylab = 'Temperatura prevista - graus CÃ©lsius($\mathrm{^o}$C)' 
fsize, lsize = 22, 10 ##tamanho

####eixo x - Horario ####
horas = np.linspace(0, 3*7, 8)
print(horas)
tempo = ['Qua 14h','Qui 16h','Qui 7h','Sab 4h','Dom 7h','Seg 10h','Ter 13h','Qua 16h']

##PrevisÃ£o de temperatura ###
####eixo y - temperatura prevista
Temp = np.array([8,7,6,9,13,14,13,12]) 

### Plotar dados ######
ax.plot(tempo, Temp, marker = 'o', lw = 2, ls = '--', ms = 10, color = 'blue') 
#marker bolinha na linha, lw= grossura da linha, ms = tamamnho da bolinha, ls pontilhado

#configurar grÃ¡fico
#ax.set_xlabel(xlab, fontsize=fsize) #3setando o label do x
ax.set_ylabel(ylab, fontsize=fsize)
ax.tick_params(axis = 'x', labelsize=lsize, width = 5, labelrotation = 30) ##espaÃ§os que tem, sÃ³ aplicou no x ou y, both Ã© o padrÃ£o, width = 15 largura do traÃ§o
ax.grid(ls = '--') ##colocando um grid ls -- deixa tracejado as linhas
        
fig.savefig(nomefig)
plt.show()
```
